```python
import pandas as pd
```

```python
df1 = pd.read_excel("/content/drive/MyDrive/Задание Excel-VBA 02.xlsx", sheet_name="Выгрузка 01",header=0, date_parser="дата"  )
df1
```

```python
df2 = pd.read_excel("/content/drive/MyDrive/Задание Excel-VBA 02.xlsx", sheet_name="Выгрузка 02",header=0, date_parser="дата"  )
df2
```

**Соединяем таблицы, что бы найти даные уникальные данные**

```python
m = (df1.merge(df2, how='outer', on=['ID записи'], 
              suffixes=['', '_new'], indicator=True))
m
```

**Новые данные в новой выгрузке**

```python
new_data = m.query("_merge=='right_only'")
new_data
```

**Данные которые исчели из новой выгрузки**

```python
missing_data = m.query("_merge=='left_only'")
missing_data
```

**Данные записи которых имеют отличия в любых полях**

```python
have_differences = df1.merge(df2, how='inner', on=['ID записи'], suffixes=['', '_new'])
have_differences
```

**Дописываем результат в анализируемый файл excel**

```python
with pd.ExcelWriter('/content/drive/MyDrive/Задание Excel-VBA 02.xlsx',mode='a') as writer:  
    new_data.to_excel(writer, sheet_name='Новое в новой выгрузке', index=False)
    missing_data.to_excel(writer, sheet_name='Отсутсвует в новой выгрузке', index=False)
    have_differences.to_excel(writer, sheet_name='Записи с расхождениями', index=False)
```
